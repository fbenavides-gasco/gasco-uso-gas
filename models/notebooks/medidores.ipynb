```python
def distancia(lat_o,lon_o,lat_d,lon_d):
    D = np.arccos(np.sin(lat_o)*np.sin(lat_d) + np.cos(lat_o)*np.cos(lat_d)*np.cos(lon_o-lon_d))
    return D
```

```python
aux_medidor = []
aux_cliente = []

for cliente in list(medidores.email[1028:]):
    if "@" in cliente: 
        print(cliente)
        direcc_aux = direcciones
        lat_d = medidores.query(f"email == '{cliente}'").lat.min()
        lon_d = medidores.query(f"email == '{cliente}'").long.min()
        direcc_aux['lat'] = lat_d
        direcc_aux['lon'] = lon_d
        direcc_aux['dist'] = direcc_aux.apply(lambda x: geopy.distance.distance((x.latitude,x.longitude), (x.lat,x.lon)).km,axis=1)
        min_dist = direcc_aux.dist.min()
        aux_cliente.append([cliente, min_dist])
    else:
        continue
clientes_final_2 = pd.DataFrame(aux_cliente, columns=['mail', 'min_distancia'])

clientes_final_2.head()
```
